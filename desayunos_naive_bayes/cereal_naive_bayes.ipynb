```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB

from google.colab import drive
drive.mount('/content/drive')


# Carga de datos
data_path = '/content/drive/MyDrive/Colab Notebooks/anilitica1/cereal2.csv'  # Asegúrate de cambiar esta ruta al lugar donde tienes tu archivo
data = pd.read_csv(data_path, sep=';')

# Preparación de datos
encoder = OrdinalEncoder()
X = data.drop('desayuno', axis=1)  # Variables predictoras
y = data['desayuno']  # Variable objetivo

# Codificación de variables categóricas
X_encoded = encoder.fit_transform(X)
y_encoded = encoder.fit_transform(y.values.reshape(-1, 1)).ravel()

# División de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Inicialización y entrenamiento del modelo Naive Bayes
model = CategoricalNB()
model.fit(X_train, y_train)

# Evaluación del modelo
accuracy = model.score(X_test, y_test)
print(f"Accuracy of the model: {accuracy*100:.2f}%")
```
